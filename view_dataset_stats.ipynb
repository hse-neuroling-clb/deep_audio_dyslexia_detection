```python
import numpy as np


def to_native(x):
    if isinstance(x, dict):
        return {k: to_native(v) for k, v in x.items()}
    if isinstance(x, list):
        return [to_native(v) for v in x]
    if isinstance(x, np.ndarray):
        return [to_native(v) for v in x.tolist()]
    if isinstance(x, (np.integer,)):
        return int(x)
    if isinstance(x, (np.floating, float)):
        f = float(x)
        return int(f) if f.is_integer() else round(f, 2)
    if isinstance(x, (np.bool_,)):
        return bool(x)
    return x


def get_participant_stats(df, text:str, grade:int|None=None):
    stats = dict()
    
    if grade is not None:
        df = df[df["grade"] == grade]
    
    stats["grade"] = grade if grade is not None else "all"

    if "task" in df.columns:
        df = df[df["task"] == text]
    elif "text" in df.columns:
        df = df[df["text"] == text]
    else:
        raise ValueError("No 'task' or 'text' column found in the dataframe.")
    stats["text"] = text

    
    boys = df[df["gender"] == 1.0]
    girls = df[df["gender"] == 0.0]

    stats[f"total"] = {
        "total": len(df),
        "boys": boys.shape[0],
        "girls": girls.shape[0],
    }
    stats["dyslexia"] = {
        "TD_girls": ((girls["label1"] == 2)).sum(),
        "TD_boys": ((boys["label1"] == 2)).sum(),
        "R_girls": ((girls["label1"] == 1)).sum(),
        "R_boys": ((boys["label1"] == 1)).sum(),
        "D_girls": ((girls["label1"] == 0)).sum(),
        "D_boys": ((boys["label1"] == 0)).sum(),
    }
    assert sum(stats["dyslexia"].values()) == df.shape[0]
    return to_native(stats)
```

```python
from typing import Literal

def make_row(stats):
    
    row = {
        "grade": stats['grade'],
        "text": stats['text']
    }
    counts = stats['dyslexia']
    row.update({
        **{k: v for k, v in counts.items()}
    })
    return row




def make_df_for_text(df, text):
    text_table = pd.DataFrame()


    for grade in GRADES:
        stats = get_participant_stats(df.copy(), text=text, grade=grade)
        participants_row = pd.DataFrame([make_row(stats)])
        participants_row["grade_count"] = participants_row[["TD_boys", "TD_girls", "R_boys", "R_girls", "D_boys", "D_girls"]].sum(axis=1)
        text_table = pd.concat([text_table, participants_row], ignore_index=True)
    
    total_row = text_table.sum().to_frame().T
    total_row["grade"] = "total"
    total_row["text"] = text

    text_table = pd.concat([total_row, text_table], ignore_index=True)

    return text_table


def format_df(df):
    def format_group_value(row, group:Literal["TD", "R", "D"]):
        boys = row[f"{group}_boys"]
        girls = row[f"{group}_girls"]
        formatted_group_value =  f"{boys+girls}\n({girls} girls, {boys} boys)"
        return formatted_group_value


    result = pd.DataFrame(columns=["Group", "TD", "R", "D"])
    for _, row in df.iterrows():
        group_row = {
            "Group": f"Grade {row['grade']}\nN={row['grade_count']}",
            **{
                group: format_group_value(row, group) for group in ["TD", "R", "D"]
            },
        }
        result = pd.concat([result, pd.DataFrame([group_row])], ignore_index=True)
    return result

```

```python


def make_exploratory_table(df):
    text_dfs = {
        text: format_df(make_df_for_text(df, text)) for text in TEXTS
    }

    final_df = pd.DataFrame()
    for grade in GRADES:
        temp_df = pd.DataFrame()
        for k, v in text_dfs.items():
            temp_df = pd.concat([temp_df, v[v["Group"].str.contains(f"Grade {grade}")].assign(Text=k)], ignore_index=True)
        final_df = pd.concat([final_df, temp_df], ignore_index=True)
    return final_df
```
